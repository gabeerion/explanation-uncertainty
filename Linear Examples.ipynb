## Display data

```python
X,y = default_data()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(X[:,0],X[:,1],c=y)
plt.title('Data Scatterplot')
plt.xlabel("$X_0$")
plt.ylabel("$X_1$")
cbar = plt.colorbar()
cbar.set_label("Y")
```

## Sklearn Linear Model

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
LR = LinearRegression()
LR.fit(X,y)
yp = LR.predict(X)
print(f'Sklearn $R^2={r2_score(y,yp):.3f}$')
print(f'Sklearn $\beta_0={LR.coef_[0]:.3f}$')
print(f'Sklearn $\beta_1={LR.coef_[1]:.3f}$')
print(f'Sklearn Intercept={LR.intercept_:.3f}')
```

```python
plt.figure(figsize=(6,6))
plt.scatter(y,yp)
plt.title(f'Sklearn $R^2={r2_score(y,yp):.3f}$')
plt.xlabel("True $y$")
plt.ylabel("Predicted $\hat{y}$")
```

## Torch Linear Model

```python
from torch_linear import TorchLinear
```

```python
TL = TorchLinear(lr=0.003)
TL.fit(X,y)
yp = TL.predict(X)
print(f'Torch $R^2={r2_score(y,yp):.3f}$')
print(f'Torch $\beta_0={TL.coef_[0]:.3f}$')
print(f'Torch $\beta_1={TL.coef_[1]:.3f}$')
print(f'Torch Intercept={TL.intercept_:.3f}')
```

```python
plt.figure(figsize=(6,6))
plt.scatter(y,yp)
plt.title(f'Torch $R^2={r2_score(y,yp):.3f}$')
plt.xlabel("True $y$")
plt.ylabel("Predicted $\hat{y}$")
```
